## Properties with time code

```python
import pandas as pd

csv_files = ['properties_1_to_50.csv', 'properties_51_to_100.csv', 'properties_100_to_200.csv','properties_200_to_300.csv', 
             'properties_300_to_400.csv', 'properties_400_to_600.csv', 'properties_600_to_800.csv', 'properties_800_to_1000.csv', 
             'properties_1000_to_1200.csv', 'properties_1200_to_1300.csv','properties_1300_to_1400.csv' ,
             'properties_1400_to_1500.csv','properties_1500_to_1600.csv', 'properties_1600_to_1700.csv', 'properties_1701_to_1800.csv', 
             'properties_1801_to_1900.csv','properties_1900_to_2000.csv', 'properties_2001_to_2050.csv', 'properties_2061_to_2084.csv' ]  
dataframes = []

for file in csv_files:
    df = pd.read_csv(file)
    dataframes.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)
combined_df.rename(columns={'Type': 'type', 'Completion status': 'completion_status','Furnishing': 'furnishing', 
                            'Reactivated date': 'post_date', 'property_header': 'address', 'Price': 'none',  'Average Rent': 'average_rent',  
                           'Purpose': 'purpose','Trucheck date': 'year_of_completion', 'area':'area'  }, inplace=True)

combined_df.to_csv('combinedproperties.csv', index=False)

print("CSV files combined successfully into 'combinedproperties.csv'.")
```

```python
df = pd.read_csv('combinedproperties.csv')
df['address'] = df['address'].fillna('')
df['Locality'] = df['address'].apply(lambda x: x.split(',')[-2].strip() if ',' in x else '')
df['Region'] = df['address'].apply(lambda x: x.split(',')[-1].strip() if ',' in x else '')
df['purpose'] = 'for sale'
df['Country'] = 'UAE'
unwanted_columns = ['year_of_completion', 'average_rent', 'Handover date', 'none', 'Amount paid'] 
df = df.drop(columns=unwanted_columns)
df.to_csv('sale_properties.csv', index=False)

df.head()
```

```python
import pandas as pd

files = ['list100.csv', 'list200.csv', 'list300.csv', 'list400.csv', 'list500.csv',  'list500_1000.csv', 'list1000to1500.csv'  ,'list1500to2000.csv', 'list2000_2060.csv', 'list2061to2084.csv', ] 

dataframe = []

for file in files:
    df_bayut = pd.read_csv(file)
    dataframe.append(df_bayut)
#the coloumn names where mismatched while scraping
all_df= pd.concat(dataframe, ignore_index=True)
all_df.rename(columns={'Type': 'type', 'Completion status': 'completion_status','Furnishing': 'furnishing', 'Reactivated date': 'post_date', 'property_header': 'address', 'Price': 'none',  'Average Rent': 'average_rent',  
                    'Property TruCheck verification date': 'year_of_completion', 'Area':'area', 'Property description': 'description', 'Name':'title', 'bath':'baths', 'Agency name': 'agency_name' }, inplace=True)
all_df['purpose'] = 'for sale'
all_df.to_csv('sale_list.csv', index=False)

print("CSV files combined successfully into 'sale_list.csv'.")
```
